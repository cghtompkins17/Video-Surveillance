```python
import os
import cv2
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
folder="C:/Users/CGH Tompkins/Desktop/View_001"
```

```python
img_list = load_images_from_folder(folder)
```

```python
from matplotlib import pyplot as plt
cnt = list()
count=0
```

```python
for i in range(2):
    fig = plt.figure(figsize=(12,12))
    plt.imshow(img_list[count])
    plt.show()
    cnt.append(int(input()))
```
