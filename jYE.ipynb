```python
import os
import cv2
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
folder="C:/Users/CGH Tompkins/Desktop/View_001"
img_list = load_images_from_folder(folder)
from matplotlib import pyplot as plt
cnt = list()
count=0
for i in range(len(img_list)):
    fig = plt.figure(figsize=(12,12))
    plt.imshow(img_list[count])
    plt.show()
    cnt.append(int(input("Enter number")))
    clear_output()
    count += 1
```

```python
import pandas as pd
category = pd.read_csv('C:/Users/CGH Tompkins/Desktop/Kolekar.csv')
```

```python
for i in cnt:
    if i <= 9 :
        category = category.append({'Category' : 'VL'}, ignore_index=True)
    elif i<=16:
        category = category.append({'Category' : 'L'}, ignore_index=True)
    elif i<=24:
        category = category.append({'Category' : 'M'}, ignore_index=True)
    elif i <=32:
        category = category.append({'Category' : 'H'}, ignore_index=True)
    else:
        category = category.append({'Category' : 'VH'}, ignore_index=True)
```

```python
category.to_csv('Category.csv')
```
